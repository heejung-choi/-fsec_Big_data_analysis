## 1. 결측값 확인 및 처리
 - pd.isnull, pd.dropna, pd.fillna

```python
# 컬럼이름을 지정해주었다.
df = pd.DataFrame(np.random.randn(5,5), columns=['V','W','X','Y','Z'])
# 조건식 적용
df_nan = df[df>-1]
df_nan
```

```python
# 결측값 존재 유무 확인
df_nan.isnull()
```

```python
# 결측값 제거(row)
df_nan.dropna()
```

```python
df_nan.dropna(axis=1)
```

```python
# 특정 열의 결측값만 처리
df_nan.dropna(axis = 0, subset = ['Z'], inplace=True)
# inplace 원본을 대체한다.
df_nan
```

```python
# 결측값을 특정 값으로 대체
df_nan.fillna(0)
```

```python
df_nan.fillna(df_nan.mean())
```

```python
df_nan
```

## 3. 데이터 변환
 - 데이터 자료형을 적절한 형태로 변환
 - 하나의 열이 여러 개의 정보를 가지고 있을 경우 열을 분리
 - 특정 데이터의 편향도가 클 경우 데이터 분포를 정규 분포로 변형

```python
date = pd.date_range(start='2018-04-24', end='2018-04-27', periods=10)
date
```

```python
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
df['date'] = date
df
```

```python
# 업무 시간과 비 업무 시간을 구분해 새로운 열로 만들어 봅시다!

df['Workhour'] = False
# 전체를 False로 지정을 해둔다음
df
```

```python
#dataframe에 구분
for idx, row in df.iterrows():
    cond = (row['date'].hour >= 9) &  (row['date'].hour < 18 )
    df.loc[idx, 'Workhour'] = cond
    
df   
```

### 백분위 수

- quantile을 이용하여 원하는 백분위수 값을 구할 수 있다.

```python
df = pd.DataFrame(np.random.rand(1000, 5), columns=['A', 'B', 'C', 'D', 'E'])
df
```

```python
# 하위 10%
low = df['A'].quantile(0.1)
low
```

```python
# 상위 10%
high = df['A'].quantile(0.9)
high
```

```python
df = df[df['A'] > low]
df.shape
```

```python
df = df[df['A'] < high]
df.shape
```

#### BOXPLOT 구하기

```python
df = pd.DataFrame([1, 2, 3, 4, 5,5,5,5,5,5, 6, 7, 8, 9, 10], columns=['A'])
df
```

```python
df.boxplot()
```

```python
Q1 = df.quantile(.25)
print("Q1 =",Q1)
```

```python
print("Q1 =",Q1)

Q2 = df.quantile(.5)
print("Q2 =",Q2)


Q3 = df.quantile(.75)
print("Q3 =",Q3)

IQR = Q3 - Q1
print("IQR =",IQR)


Q4 = df.quantile(1)
print("Q4 =",Q4)
```

```python
df.describe()
```

```python
df = df[df['A'] > 1.5]
df = df[df['A'] < 9.5]
df.boxplot()
```

```python
# outlier

outlier_low = Q1 - (1.5 * IQR)
outlier_low
```

```python
outlier_high = Q3 + (1.5 * IQR)
outlier_high
```
